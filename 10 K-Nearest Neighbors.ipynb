```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pickle
input = open('data_class.pkl', 'rb')
obj = pickle.load(input)
input.close()
X_train = obj["X_train"]
X_test = obj["X_test"]
y_train = obj["y_train"]
y_test = obj["y_test"]
```

```python
from sklearn.neighbors import KNeighborsClassifier
error_rate = []
for i in range(20,60):
     knn = KNeighborsClassifier(n_neighbors=i)
     knn.fit(X_train,y_train)
     pred_i = knn.predict(X_test)
     error_rate.append(np.mean(pred_i != y_test))

plt.figure(figsize=(10,6))
plt.plot(range(20,60),error_rate,color='blue', linestyle='dashed', 
         marker='o',markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
print("Minimum error:-",min(error_rate),"at K =",error_rate.index(min(error_rate)))
plt.show()
```

Примерное подходящее число количества соседей было равным 37, так как это корень из количества элементов в выборке. Поэтому для данной функции был взят интервал предполагаемого количества соседей (20,60). Исходя из результатов, оптимальное количество соседей равняется 22-24 либо 32. При проведении ряда эксперементов с разными вариантами к-ва соседних получалось, точность при всех значениях одинаковая, поэтому нет никакого смысла усложнять модель. ПВ итоге количесто К-ближайших соседей равно 22. 

```python
knn = KNeighborsClassifier(n_neighbors = 22, metric = 'minkowski', p = 2).fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
knn.score(X_test,y_test)
```

Точность данной модели классификации немного лучше логистической регрессии.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

Также исследуемая модель немного лучше справилась с рапознаванием элементов, так как количество неверно распознанных элементов стало 185, у логистической модели оно было равно 187. Точность модели чуть выше логистической модели. То есть в целом данная модель лучше.
