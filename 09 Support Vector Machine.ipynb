```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pickle
input = open('data_class.pkl', 'rb')
obj = pickle.load(input)
input.close()
X_train = obj["X_train"]
X_test = obj["X_test"]
y_train = obj["y_train"]
y_test = obj["y_test"]
```

```python
from sklearn.svm import SVC
svm = SVC(kernel = 'rbf', random_state = 10).fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)
svm.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
svm = SVC(kernel = 'linear', random_state = 10).fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)
svm.score(X_test,y_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
svm = SVC(kernel = 'poly', random_state = 10).fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)
svm.score(X_test,y_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
svm = SVC(kernel = 'sigmoid', random_state = 10).fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_test)
svm.score(X_test,y_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

Модель на основе сигмоидального ядра имееет самое высокое значение точности, но классификация хуже всего распознает объекты. Таким образом наилучшей моделью является модель на основе линейного опорного вектора. Её результаты - это 187 неверно распознаных объектов. Коэффициент детерминации такой же, как и у логистической модели.
