```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pickle
input = open('data_class.pkl', 'rb')
obj = pickle.load(input)
input.close()
X_train = obj["X_train"]
X_test = obj["X_test"]
y_train = obj["y_train"]
y_test = obj["y_test"]
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB().fit(X_train, y_train)
```

```python
y_pred = nb.predict(X_test)
nb.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

Наивный байесовский классификатор справился с классификацией хуже, нежели это было в логистической регрессии. Но стоит отметить, что этот метод, по сравнению со всеми остальными, лучше распознает негативные случаи, то есть тех клиентов, у которых отсутствует привязанность к сайту. Но, так как уровень точности классификации намного ниже и процент объясненности меньше, чем у остальных моделей, данный способ классификации оказывается в аутсайдерах.
